```python
df = pd.read_csv('../input/disease-symptom-description-dataset/dataset.csv')
df.head()
```

```python
df.Symptom_1.isnull().sum()
```

# **Symptoms Dummies**

```python
symptom_dummies = pd.get_dummies(df.iloc[:,1:])
symptom_dummies.head()
```

```python
symptom_dummies_cols = pd.get_dummies(df.iloc[:,1:]).columns
symptom_dummies_cols
```

# **All Symptoms Columns of OG Dataset**

```python
symptom_cols = df.iloc[:,1:].columns
symptom_cols
```

# **Finding All Unique Symptoms**

```python
unique_symptoms = []

for i in df.columns[1:] : 
    for j in df[i].unique():
        if j not in unique_symptoms and type(j) == str: 
            unique_symptoms.append(j)
```

```python
unique_symptoms = np.array(unique_symptoms)
unique_symptoms[0] = ' itching'
unique_symptoms
```

```python
# arrenging symptoms in ascending order

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(unique_symptoms)
encoder.classes_
```

# **Creating DF2**

```python
symptom_dict = {i:[] for i in encoder.classes_}
symptom_dict
```

```python
for i in encoder.classes_:
    
    temp = [0 for i in range(df.shape[0])]
    for j in symptom_cols:
        
        col_str = j+'_'+i
        if col_str in symptom_dummies_cols:
            
            temp += symptom_dummies[j+'_'+i]
            
    symptom_dict[i] = np.array(temp)
```

```python
df2 = pd.concat([df[['Disease']],pd.DataFrame(symptom_dict)],axis = 1)
df2.shape
```
